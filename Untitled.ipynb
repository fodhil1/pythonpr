```python
# We need to import the libraries necessary to store our data and visualise our data
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
#import modules
df_tes = pd.read_csv('popul.csv')
con = sqlite3.connect('tes.db')
df_tab = df_tes.to_sql('tab_tes', con, if_exists ="replace")

df_tes
```

```python
# retrive the data we need to track the evolution of the population

USA_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'USA'", con)

#USA_df

BRA_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'BRA'", con)
#BRA_df

RUS_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'RUS'", con)
#RUS_df

GBR_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'GBR'", con)

#GBR_df

CHN_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'CHN'", con)

#CHN_df

DEU_df = pd.read_sql_query("SELECT Value FROM tab_tes WHERE [Country Code] = 'DEU'", con)

#DEU_df

year = pd.read_sql_query("SELECT Year FROM tab_tes WHERE [Country Code] = 'DEU'", con)

#year
```

```python
# Here we plot the graphs 

plt.plot(year, USA_df, color = 'green', label='USA')
plt.plot(year, GBR_df, color = 'red', label='GBR')
plt.plot(year, BRA_df, color = 'black', label='BRA')
plt.plot(year, RUS_df, color = 'blue', label='RUS')
#plt.plot(year, CHN_df, color = 'yellow', label='CHN')
plt.plot(year, DEU_df, color ='orange', label='DEU')
plt.xlabel('Year')
plt.ylabel('Population in million')
plt.title('Evolution of Population from 1960 to 2016')
plt.legend(loc='upper left')
plt.show()
```

From the graph, we conclude that the population of Great Braitin, Germany, and Russia is stayed constent, but the population of USA and Brazil increased significantly from 1960 to 2016.
